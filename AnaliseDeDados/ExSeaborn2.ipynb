```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
esc_x = [1989, 1997, 2002, 2007, 2012]
esc_y = np.arange(0,201,50)

fig, ax = plt.subplots(figsize=(10, 5))
plt.ylabel("count")
plt.xticks(esc_x)
plt.yticks(esc_y)

sns.set_theme(style="white")
planets = sns.load_dataset("planets")
sns.histplot(
    data=planets,
    x="year",
    discrete = True
    )
```
