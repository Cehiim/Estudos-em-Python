```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

media = 0.1
x = np.arange(0, 6)
y = poisson.pmf(x, media)

plt.plot(x, y, 'o-')
plt.xlabel('Eventos')
plt.ylabel('Valor')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

media = 2
x = np.arange(0, 11)
y = expon.pdf(x, scale=media)

plt.plot(x, y, 'b-')
plt.xlabel('Valor')
plt.ylabel('Probabilidade')
plt.show()
```
