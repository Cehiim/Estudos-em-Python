```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
fig, ax = plt.subplots()

ling = ('Python','C','C++','C#','Java','SQL','JavaScript','R','HTML','TypeScript')
uso = [100,96.8,88.58,86.99,70.22,47.37,40.48,18.92,17.97,16.99]
width = 0.60
lug = np.arange(len(ling))

for i, v in enumerate(uso):
    ax.text(v - 10, i + .1125, str(v), color='white', fontweight='medium', font='sans serif')
ax.barh(lug, uso, width, color = 'tab:orange')
ax.set_yticks(lug, labels=ling)
ax.invert_yaxis()
plt.tick_params(left = False, bottom = False)
ax.set_xticklabels('')
ax.set_facecolor('lightgray')

plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

Notas_Matematica = [88, 92, 80, 89, 100, 80, 60,100, 80, 34]
Notas_Ciencias = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]
avaliacao = np.arange(10,110,10)



fig, ax = plt.subplots()
ax.scatter(avaliacao, Notas_Matematica,marker="v",color="red",label="Notas Matematica")
ax.scatter(avaliacao, Notas_Ciencias,marker=".",color="blue",label="Notas Ciencia")
plt.xlabel('Avaliacao')
plt.ylabel('Notas')
ax.legend(loc = "lower left")
plt.xlim([0,120])
plt.show()
```
